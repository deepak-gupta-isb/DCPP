```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
# Importing Dataset
```

```python
df = pd.read_excel('amazon_scraped_processed.xlsx')
```

```python
df.head(5)
```

```python
df.info()
```

```python
# Data Cleaning
```

```python
# Converting Reviews column from String to Float 
```

```python
rat = []
for x in df['rating'].values:
    try:
        z = x[:3]
        rat.append(float(z))
    except TypeError as er:
        rat.append(0)
```

```python
df1 = df.copy()
```

```python
df1['rating'] = rat
```

```python
df1[df1['rating']==0]
```

```python
df1['review'].fillna(0, inplace = True)
```

```python
# Removing irrelevant data from stock_tag column
```

```python
stock = []
for x in df1['stock_tag'].values:
    try:
        if('stock' in x)==True:
            stock.append(x)
        else:
            stock.append(np.nan)
    except TypeError as er:        
            stock.append(np.nan)
```

```python
df1['stock_tag']= stock
```

```python
# Rounding off discount column to 2 decimal places
```

```python
df1['discount'] = round(df1['discount'],2)
```

```python
df1.isnull().sum()
```

```python
# Replacing 0s with NaN 
```

```python
df1['review'].replace(0,np.nan, inplace = True)
```

```python
df1['rating'].replace(0,np.nan, inplace = True)
```
